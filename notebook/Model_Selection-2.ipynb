# SPLIT TRAIN AND TEST DATA

```python
X_train = train_df.drop('price', axis=1)
y_train = train_df['price']

X_test = test_df.drop('price', axis=1)
y_test = test_df['price']
```

# LINEAR REGRESSION

```python
'''
import mlflow.sklearn
import mlflow.sklearn
'''
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

SS = StandardScaler()
X_train_scaled = SS.fit_transform(X_train)
X_test_scaled = SS.transform(X_test)

y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
LR = LinearRegression()
RD = Ridge()
LS = Lasso()
EN = ElasticNet()

models = [LR, RD, LS, EN]

for model in models:
  model.fit(X_train_scaled, y_train_log)
  y_pred_log = model.predict(X_test_scaled)
  y_pred = np.expm1(y_pred_log)


  print(str(model))
  print(f'mean absolute error -> {mean_absolute_error(y_test, y_pred)}')
  print(f'mean squared error -> {mean_squared_error(y_test, y_pred)}')
  print(f'r2 score -> {r2_score(y_test, y_pred)}')
  print("")
  print("----------------------")
```

# XGBREGRESSOR

```python
from xgboost import XGBRegressor
```

```python
model_xgb = XGBRegressor()
model.fit(X_train, y_train_log)
y_pred = model.predict(X_test)
y_pred = np.expm1(y_pred)

print(f'XGBRegressor model')
print(f'mean absolute error -> {mean_absolute_error(y_test, y_pred)}')
print(f'mean squared error -> {mean_squared_error(y_test, y_pred)}')
print(f'r2 score -> {r2_score(y_test, y_pred)}')
print("")
print("----------------------")
```

## XGBoost performed the best overall. It required less time and fewer computational resources, and it achieved better performance than TensorFlow. The TensorFlow model recorded an MAE of 53,935.74, whereas the XGBoost model achieved a lower MAE of 41,548.56, indicating superior predictive accuracy.

# TENSORFLOW MODEL

```python
X_train_scaled.shape , X_test_scaled.shape
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential, Input
from keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping

SS = StandardScaler()
X_train_scaled = SS.fit_transform(X_train)
X_test_scaled = SS.transform(X_test)

y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)


model = Sequential([
    Input(shape=(32,)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1)
])
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

from sklearn.model_selection import train_test_split
X_tr , X_val, y_tr, y_val = train_test_split(X_train_scaled, y_train_log, test_size=0.2, random_state=42)

early_stopping = EarlyStopping(monitor='val_mae', patience=5, restore_best_weights=True)


model.fit(X_tr, y_tr,
          epochs=300,
          validation_data=(X_val, y_val),
          callbacks=[early_stopping])

y_pred = model.predict(X_test_scaled)
y_pred = np.expm1(y_pred)
y_pred = y_pred.ravel()

mae = mean_absolute_error(y_test, y_pred)

print('validation mae : ', mae)
```

# MODEL ANALYSIS

XGBoost performed the best overall. It required less time and fewer computational resources, and it achieved better performance than TensorFlow.
The TensorFlow model recorded an MAE of 53,935.74, whereas the XGBoost model achieved a lower MAE of 41,548.56, indicating superior predictive accuracy.
